```python
def padovan(n):
    Padovan = [1, 1, 1]
    if n <= 3:
        return Padovan
    else:
        for i in range(3, n):
            NextNum = Padovan[i - 2] + Padovan[i - 3]
            Padovan.append(NextNum)
        return Padovan

n = int(input('Please enter the list length of Padovan sequences: '))
PadovanNums = padovan(n)
print('Padovan sequence: ')
print(PadovanNums)
```
